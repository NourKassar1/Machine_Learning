### **F[i]=wx[i]+b **






```python
#This function to compute the expected value
def compute_modle_output(x,w,b):
  l=x_train.shape[0]
  f=np.zeros(l)
  for i in range(l):
    f[i]=w*x_train[i]+b
  return f
```



  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 \tag{1}$$ 





```python
#Calculate the cost function : J(w,b)
def cost_function(x,w,b,y):
  len=x.shape[0]
  cost=0
  for i in range(len):
    f[i]=w*x[i]+b
    error=(f[i]-y[i])**2
    cost+=(1/(2*len))*error
  return cost  
```

```python
f=compute_modle_output(x_train,w,b)
cost=cost_function(x_train,w,b,y_train)
print(cost)
```

```python
# Plot our model prediction
plt.plot(x_train, f, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

plt.legend()
plt.show()
```
